```python
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
df['release_date'] = df['release_date'].dt.year
```

```python
for i in df.index:
    df.loc[i, 'release_date'] = str(int(df.loc[i, 'release_date']/10)) + '0'
```

```python
df.sample(5)
```

```python
df['overview'] = df['overview'].str.split(",")
df['genres'] = df['genres'].str.split(",")
df['writers'] = df['writers'].str.split(",")
df['producers'] = df['producers'].str.split(",")
df['cast'] = df['cast'].str.split(",")
```

```python
df['director'] = df['director'].str.split(",")
df['release_date'] = df['release_date'].str.split(",")
```

```python
df.head()
```

```python
df['overview'] = df['overview'].apply(lambda x:[i.replace(" ",",") for i in x])
```

```python
df['genres'] = df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast'] = df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['director'] = df['director'].apply(lambda x:[i.replace(" ","") for i in x])
df['writers'] = df['writers'].apply(lambda x:[i.replace(" ","") for i in x])
df['producers'] = df['producers'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df.head()
```

```python
df['tags'] = df['overview'] + df['genres'] + df['cast'] + df['director'] + df['writers'] + df['producers'] + df['release_date']
```

```python
df = df[['id','title','tags']]
```

```python
df.head()
```

```python
df=df.reset_index()
```

```python
df['tags'] = df['tags'].apply(lambda x:" ".join(x))
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vector = cv.fit_transform(df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity[0]
```

```python
def recommend(movie):
    index = df[df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])[1:6]
    for i in distances:
        print(df.iloc[i[0]].title)
        print(df.iloc[i[0]].id)

```

```python
recommend('Avatar')
```
